```python
import numpy as np
v = np.array([1,-1, 2])
w = np.array([2,5,2])
```

```python
print(v+w)

[3,4,4]
```

```python
print(v-w)

[-1, -6, 0]
```

```python
#Scalar Multiplication
print(3*v)

[2, -2, 4]
```

```python
#length
print(np.linalg.norm(v))
```

```python
#dot product
s = np.dot(v,w)
print(s)
```

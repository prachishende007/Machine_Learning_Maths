```python
import numpy as np
A=np.array([[2, -2, 3], [1, 1, 1], [1, 3, -1]])
print(A)
```

```python
#calculate eigenpairs :
values, vectors = np.linalg.eig(A)
print(values)
print(vectors)
```

```python
import numpy as np
A=np.array([[2, 0, 0], [0, 1, 0], [0, 0, -1]])
print(A)
```

```python
#calculate eigenpairs :
values, vectors = np.linalg.eig(A)
print(values)
print(vectors)
```

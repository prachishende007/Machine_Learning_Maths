```python
import numpy as np
p = np.array([[1,7],[2,1],[3,2]])
print(p)
```

```python
q = np.array([[1,2],[4,5],[2,3]])
print(q)
```

```python
c1 = np.add(p,q)
print(c1)
```

```python
c2 = np.subtract(p,q)
print(c2)
```

```python
r = np.array([[1,3,1],[1,0,1]])
print(r)
```

```python
c3 = np.dot(p,r)
print(c3)
```

```python
determinant = np.linalg.det(c3)
print(determinant)
```

```python
m = np.array([[1,1],[2,3]])
inverse = np.linalg.inv(m)
print(inverse)
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.random.rand(50,1)
Y = 1.3 * (X**8) - 1.4 * (X**7) - 0.8 * (X**6) + 0.6 * (X**5) + 0.3 * (X**4) + 0.5 * (X**3) + 1.2 * (X**2) - 0.6 * X + 0.1 * np.random.rand(50, 1)
```

```python
print(X)
```

```python
fig = plt.figure()
plt.scatter(X, Y)
plt.show()
```

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
```

```python
for i in range(1, 6):
    poly = PolynomialFeatures(i) # degree of features
    x_poly = poly.fit_transform(X)
    scaler = StandardScaler() # standardization between 0-1
    scaler.fit(x_poly)
    x_poly_stnd = scaler.transform(x_poly)
    model = LinearRegression()
    model.fit(x_poly_stnd, Y)
    output = model.predict(x_poly_stnd)
    fig = plt.figure()
    plt.scatter(X, output, label = 'predict')
    plt.scatter(X, Y, label = 'given')
    plt.show()
```

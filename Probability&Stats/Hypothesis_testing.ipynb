```python
import matplotlib.pyplot as plt
import scipy.stats as stat
import numpy as np
```

Null Hypothesis Testing

```python
n_patients = 1000
p_success = 0.9
dist_binomial = stat.binom(n = n_patients, p = p_success)
x_values = np.arange(n_patients + 1, step = 5)
y_values = dist_binomial.pmf(x_values)
plt.stem(x_values, y_values)
```
